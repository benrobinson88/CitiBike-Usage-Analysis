```python
#Digging into Trip Duration
#First, let's (arbitrarily) replace any trips over 90 minutes (5400 seconds) with np.nan
df['tripduration'].replace(np.arange(5401,6251000), np.nan)
df['tripduration'].max()

#In theory the above seems to work (tested with number much smaller than 6251000), but super slow.
#One option:  convert seconds to minutes, then do it.
#Better option:  see below
```

```python
#Has to be a faster way:
df1 = df[df['tripduration']<=5400]
df2 = df[df['tripduration']>5400]
df2['tripduration'] = np.nan
frames = [df1, df2]
df = pd.concat(frames)
df['tripduration'].max()
```

```python
#Convert seconds to minutes
df['tripduration']=df['tripduration']/60
```
